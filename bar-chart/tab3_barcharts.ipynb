### Tab 3: Bart Charts moving over time:

- Two bart charts on the same line with the bars moving as the years go by.
- A first bart chart to show the evolution of the countries ranking in terms of number of wins or points or players etc...
- A second bart chart to show the evolution of the players in terms of number of wins or points or number of matches etc...
- Button to apply filters according to different variables: Series, Court, Surface.
- Inspiration: https://observablehq.com/@d3/bar-chart-race

```python
# Open data
df = pd.read_csv("../data/tennis_data_cleaned.csv", index_col=0)
df.head()
```

```python
def lil_func(x):
    x = x[:5]+'01-01'
    return x
df.Date = df.Date.apply(lambda x: lil_func(x))
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
df
```

```python
df.columns
```

```python
winners = pd.Series(df['Winner'].unique())
losers = pd.Series(df['Loser'].unique())
all_players = winners.append(losers)
all_players = pd.Series(all_players.unique())
all_players
```

```python
barchart1 = pd.DataFrame(columns=['date','name','value'])
barchart1
```

```python
for_los = df.groupby(['Date','Loser']).sum()['LPts'].reset_index().rename(columns={"Date":'date',"Loser": "name", 'LPts':'value'})
#print(for_los.head(-5))
counts_of_match_losers = df.groupby(['Date','Loser']).count()['LPts'].reset_index().LPts
#print(df.groupby(['Date','Loser']).count()['LPts'].reset_index().head(-5))
for_los.value = for_los.value / counts_of_match_losers
for_los.head(-5)
```

```python
for_win = df.groupby(['Date','Winner']).sum()['WPts'].reset_index().rename(columns={"Date":'date',"Winner": "name", 'WPts':'value'})
#print(for_win.head(-5))
counts_of_match_winners = df.groupby(['Date','Winner']).count()['WPts'].reset_index().WPts
#print(df.groupby(['Date','Winner']).count()['WPts'].reset_index().head(-5))
for_win.value = for_win.value / counts_of_match_winners
for_win.head(-5)
```

```python
barchart1 = pd.concat([for_los, for_win])
barchart1
```

```python
barchart1.groupby(['date','name']).max().reset_index()
```

```python
barchart1 = barchart1
```

```python
barchart1.to_csv('../data/barchart1.csv')
```

```python
barchart2 = df[(df.Date.dt.year > 2007) & (df.Date.dt.year < 2023)].groupby(['Date', 'pl1_flag']).count().reset_index().iloc[:,:3]
barchart2 = barchart2.rename(columns={'Date':'date', 'pl1_flag':'name','ATP':'value'})
barchart2['category'] = barchart2.name
barchart2
```
